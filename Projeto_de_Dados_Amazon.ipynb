```python
import pandas as pd
import numpy as np
import seaborn as sns

Amazon_Datas = pd.read_csv('amazon.csv')

Amazon_df = pd.DataFrame(Amazon_Datas)

#formatando as colunas de preço, porcentagem e avaliação de string para float (numeros).
Amazon_df['discounted_price'] = Amazon_df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
Amazon_df['actual_price'] = Amazon_df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
Amazon_df['rating_count'] = Amazon_df['rating_count'].str.replace(',', '').astype(float)
Amazon_df['rating'] = pd.to_numeric(Amazon_df['rating'].str.replace('|', ''), errors='coerce')
Amazon_df['discount_percentage'] = Amazon_df['discount_percentage'].str.replace('%', '').astype(float)

#separando a coluna e guardando apenas o primeiro termo.
Amazon_df['principal_category'] = Amazon_df['category'].str.split('|').str[0]

#excluindo a coluna de categorias desnecessária.
Amazon_df.drop('category', axis=1, inplace=True)

#visualizamdo o resultado
Amazon_df.head()

Amazon_df.to_csv('Dados_Limpos_Da_Amazon.csv', index=False)
```
